```python
import pandas as pd
```

```python
country_list = pd.read_csv('data/country_list.csv')
df_globe = pd.read_csv('data/df_globe.csv')
df_flag = pd.read_csv('data/df_flag.csv')
df_info = pd.read_csv('data/df_info.csv')
df_level = pd.read_csv('data/df_level.csv')
```

```python
df_level.head()
```

```python
country_list = country_list.drop(columns='Unnamed: 0')
```

```python
df_globe = df_globe.drop(columns='Unnamed: 0')
```

```python
df_flag = df_flag.drop(columns='Unnamed: 0')
```

```python
df_info = df_info.drop(columns='Unnamed: 0')
```

```python
df_level = df_level.drop(columns='Unnamed: 0')
```

```python
df = country_list.merge(df_globe, how='left', on='country')
```

```python
df = df.merge(df_flag, how='left', on='country')
```

```python
df = df.merge(df_info, how='left', on='country')
```

```python
df = df.merge(df_level, how ='left', on ='country')
```

```python
df.head()
```

```python
df = df.rename(columns = {'image_name':'image_flag'})
```
